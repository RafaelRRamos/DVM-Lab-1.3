```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('used_cars.csv')

df.head()
```

```python
df.columns
```

Dataset Description

This dataset contains information about used vehicles including brand, model, model year, mileage, fuel type, transmission type, accident history, title status, and vehicle price. The dataset appears to have been collected from used vehicle listings and provides insight into factors that influence vehicle values.

I selected this dataset because vehicle pricing is affected by many variables including age, mileage, accident history, and vehicle brand. The dataset provides an opportunity to use Pandas and visualizations to identify trends in the used car market.


```python
top_brands = df['brand'].value_counts().head(10)

top_brands.plot(kind='bar', figsize=(8,5))

plt.title('Top 10 Vehicle Brands')
plt.xlabel('Brand')
plt.ylabel('Number of Vehicles')

plt.show()

top_brands
```

Question 1: Which vehicle brands appear most often in the dataset?

Answer:

The chart shows the ten most common vehicle brands in the dataset. These brands make up a large portion of the available used vehicles and provide insight into which manufacturers are most frequently represented in the market.


```python
df['price'].hist(bins=20, figsize=(8,5))

plt.title('Distribution of Vehicle Prices')
plt.xlabel('Price')
plt.ylabel('Count')

plt.show()

df['price'].describe()
```

Question 2: What is the distribution of vehicle prices?

Answer:

The histogram shows that vehicle prices vary considerably across the dataset. Most vehicles fall within the lower to middle price ranges, while fewer vehicles are listed at very high prices.


```python
df.groupby('fuel_type')['price'].mean().sort_values().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')

plt.show()
```

Question 3: How does average price differ by fuel type?

Answer:

The chart compares average vehicle prices by fuel type. Different fuel technologies can influence value, and the results show which fuel types tend to command higher prices in the used vehicle market.


```python
plt.scatter(df['mileage'], df['price'])

plt.title('Mileage vs Price')
plt.xlabel('Mileage')
plt.ylabel('Price')

plt.show()
```

Question 4: What relationship exists between mileage and price?

Answer:

The scatterplot shows how mileage and price are related. In general, vehicles with higher mileage tend to have lower prices, although there are some exceptions depending on vehicle type and condition.


```python
df.groupby('model_year')['price'].mean().tail(15).plot(
    kind='line',
    figsize=(8,5)
)

plt.title('Average Price by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Price')

plt.show()
```

Question 5: How does vehicle age affect price?

Answer:

The chart shows average vehicle prices by model year. Newer vehicles generally maintain higher values than older vehicles, demonstrating the impact of age and depreciation on used vehicle pricing.
