```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('full_grouped.csv')

df.head()
```

Dataset Description

This dataset contains worldwide COVID-19 case information collected by country over time. It includes the date, country or region, confirmed cases, deaths, recovered cases, active cases, and WHO region. Because the dataset contains both geographic (country) information and time-series (date) information, it is appropriate for analyzing trends across space and time.


```python
us = df[df['Country/Region'] == 'US']

plt.figure(figsize=(12,5))

plt.plot(pd.to_datetime(us['Date']), us['Confirmed'])

plt.title('Confirmed COVID-19 Cases in the United States Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

Visualization 1 Interpretation

The number of confirmed COVID-19 cases in the United States increased significantly over time. The trend demonstrates rapid growth throughout the pandemic period represented in the dataset.


```python
latest = df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
latest.plot(kind='bar')

plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')

plt.show()
```

Visualization 2 Interpretation

The visualization compares the countries with the highest number of confirmed COVID-19 cases. The data shows that the pandemic affected countries differently, with several countries experiencing much larger outbreaks than others.


```python
top3 = df[df['Country/Region'].isin(['US', 'Brazil', 'India'])]

pivot = top3.pivot(index='Date',
                   columns='Country/Region',
                   values='Confirmed')

pivot.plot(figsize=(10,5))

plt.title('Confirmed COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')

plt.xticks(rotation=45)

plt.show()
```

Visualization 3 Interpretation

This visualization compares confirmed COVID-19 cases over time for the United States, Brazil, and India. Although all three countries experienced increasing case counts, the growth rates differed, illustrating how the pandemic progressed differently across geographic regions.
