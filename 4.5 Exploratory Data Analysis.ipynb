4.5 EDA and Preparing Data for Machine Learning Models

Dataset Description

The dataset used for this project is **Employee_productivity.csv**. It contains information about employee productivity and workplace characteristics collected from a mid-sized company. The dataset includes both numerical and categorical features that may influence employee productivity.

Features

* HoursWorked – Number of hours worked.
* FocusScore – Employee focus rating.
* YearsAtCompany – Employee tenure.
* TeamMeetingsPerWeek – Number of meetings attended each week.
* DeskLocation – Employee workspace location.
* Department – Employee department.
* ProductivityScore – Overall productivity score (Target Feature).
Target Feature

The target feature for this project is **ProductivityScore** because it represents the value that we want to better understand and eventually predict using the other features in the dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Employee_productivity.csv")

df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['FocusScore'], df['ProductivityScore'])

plt.title('Focus Score vs Productivity Score')
plt.xlabel('Focus Score')
plt.ylabel('Productivity Score')

plt.show()
```

Visualization 1 Interpretation

This scatterplot shows a strong positive relationship between employee focus and productivity. Employees with higher focus scores generally have higher productivity scores, making FocusScore the strongest predictor of productivity in this dataset.


```python
plt.figure(figsize=(8,5))

plt.scatter(df['HoursWorked'], df['ProductivityScore'])

plt.title('Hours Worked vs Productivity Score')
plt.xlabel('Hours Worked')
plt.ylabel('Productivity Score')

plt.show()
```

## Visualization 2 Interpretation

The relationship between hours worked and productivity is weak. Employees who worked longer hours did not consistently achieve higher productivity scores, suggesting that simply increasing work hours does not necessarily improve performance.


```python
plt.figure(figsize=(8,5))

plt.scatter(df['YearsAtCompany'], df['ProductivityScore'])

plt.title('Years at Company vs Productivity Score')
plt.xlabel('Years at Company')
plt.ylabel('Productivity Score')

plt.show()
```

Visualization 3 Interpretation

The scatterplot shows little relationship between years at the company and productivity. Employees with both short and long tenure display a wide range of productivity scores, suggesting that experience alone is not a strong predictor of productivity.


```python
plt.figure(figsize=(8,5))

plt.scatter(df['TeamMeetingsPerWeek'], df['ProductivityScore'])

plt.title('Team Meetings per Week vs Productivity Score')
plt.xlabel('Team Meetings Per Week')
plt.ylabel('Productivity Score')

plt.show()
```

Visualization 4 Interpretation

The relationship between the number of team meetings and productivity appears to be weakly negative. Employees attending more meetings generally do not show higher productivity, suggesting that excessive meetings may reduce time available for focused work.


```python
plt.figure(figsize=(8,5))

df.boxplot(column='ProductivityScore', by='DeskLocation')

plt.title('Productivity by Desk Location')
plt.suptitle('')
plt.xlabel('Desk Location')
plt.ylabel('Productivity Score')

plt.show()
```

Visualization 5 Interpretation

The boxplot compares productivity scores across different desk locations. While there may be small differences between groups, desk location does not appear to influence productivity as strongly as employee focus.


# Conclusion

The visualizations show that **FocusScore** has the strongest relationship with employee productivity and is the best predictor of the target feature. Hours worked, years at the company, team meetings, and desk location have much weaker relationships with productivity.

These results suggest that organizations should focus on improving employee concentration and minimizing unnecessary distractions rather than simply increasing work hours or relying on employee tenure. Reducing unnecessary meetings and creating environments that support focused work may lead to higher overall productivity.
