```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

```python
df.columns
```

```python
# BPM Distribution

plt.figure(figsize=(8,5))
df['bpm'].hist(bins=20)
plt.title('Distribution of BPM')
plt.xlabel('BPM')
plt.ylabel('Count')
plt.show()

print(df['bpm'].describe())
```

```python
plt.figure(figsize=(10,5))
df['key'].value_counts().plot(kind='bar')

plt.title('Distribution of Musical Keys')
plt.xlabel('Key')
plt.ylabel('Count')

plt.show()

print(df['key'].value_counts())
```

```python
plt.figure(figsize=(6,4))
df['mode'].value_counts().plot(kind='bar')

plt.title('Distribution of Mode')
plt.xlabel('Mode')
plt.ylabel('Count')

plt.show()

print(df['mode'].value_counts())
```

```python
plt.figure(figsize=(8,5))
df['danceability_%'].hist(bins=20)

plt.title('Distribution of Danceability')
plt.xlabel('Danceability %')
plt.ylabel('Count')

plt.show()

print(df['danceability_%'].describe())
```

```python
plt.figure(figsize=(8,5))

df['energy_%'].hist(bins=20)

plt.title('Distribution of Energy')
plt.xlabel('Energy %')
plt.ylabel('Count')

plt.show()

print(df['energy_%'].describe())
```

```python
plt.figure(figsize=(8,5))

df['speechiness_%'].hist(bins=20)

plt.title('Distribution of Speechiness')
plt.xlabel('Speechiness %')
plt.ylabel('Count')

plt.show()

print(df['speechiness_%'].describe())
```

```python
plt.figure(figsize=(8,5))

df['acousticness_%'].hist(bins=20)

plt.title('Distribution of Acousticness')
plt.xlabel('Acousticness %')
plt.ylabel('Count')

plt.show()

print(df['acousticness_%'].describe())
```

```python
top_artists = df['artist(s)_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_artists.plot(kind='bar')

plt.title('Top 10 Artists in Dataset')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')

plt.show()

print(top_artists)
```

```python
engineered_hits = df[
    (df['bpm'] >= 115) &
    (df['bpm'] <= 130) &
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['danceability_%'] >= 65) &
    (df['danceability_%'] <= 80) &
    (df['energy_%'] >= 60) &
    (df['energy_%'] <= 80) &
    (df['speechiness_%'] >= 4) &
    (df['speechiness_%'] <= 11)
]

engineered_hits[['track_name','artist(s)_name']]
```
